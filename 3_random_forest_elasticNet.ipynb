# Overview

In this file we train two types machine learning models:
- Random forest regressor
- ElasticNet

For better accuracy we split the datasets into:
- Goalkeeprs
- Outfield players (all players apart from goalkeepers)

Overall we train 4 models, 2 for each type of players

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv("ready_datasets/final_dataset_for_modeling.csv", index_col="Unnamed: 0")
```

```python
data = data.dropna(subset=['role', 'price'])
data= data[data.season != 2019]
data = data.drop(["match_date", "score", "status", "player_status", "id", "coach", "coach_score"], axis=1)
```

# Goal Keepers

```python
goal_keepers = data[data.role == "G"]
```

```python
train_df_g = goal_keepers[(goal_keepers.season == 2016) | (goal_keepers.season == 2017) | ((goal_keepers.season == 2018) & (goal_keepers.matchweek < 19))]
test_df_g  = goal_keepers[(goal_keepers.season == 2018) & (goal_keepers.matchweek >= 19)]
prices_g   = test_df_g[["player", "season", "matchweek", "price"]]
```

```python
y_test_g = test_df_g["final_score_next_week"]
X_test_g = test_df_g.drop(["final_score_next_week", "price"], axis=1)
```

```python
# One-hot encode the data using pandas get_dummies
train_features_g = pd.get_dummies(train_df_g)
test_features_g = pd.get_dummies(test_df_g)
```

```python
# Get missing columns in the training test to ensure same number of columns across
missing_cols = set(train_features_g) - set(test_features_g)

# Add a missing column in test set with value equal to 0
for c in missing_cols:
    test_features_g[c] = 0

    # Ensure the order of column in the test set is in the same order than in train set
test_features_g = test_features_g[train_features_g.columns]
```

```python
# Labels are the values we want to predict
train_labels_g = np.array(train_features_g["final_score_next_week"])
test_labels_g= np.array(test_features_g["final_score_next_week"])

# Remove the labels from the features
train_features_g= train_features_g.drop(columns=["final_score_next_week", "price"], axis = 1)
test_features_g=  test_features_g.drop(columns=["final_score_next_week", "price"], axis = 1)

# Saving feature names for later use
feature_list_g = list(train_features_g.columns)

# Convert to numpy array
features_g = np.array(train_features_g)
```

```python
print('Training Features Shape:', train_features_g.shape)
print('Training Labels Shape:', train_labels_g.shape)
print('Testing Features Shape:', test_features_g.shape)
print('Testing Labels Shape:', test_labels_g.shape)
```

### Random Forrest model

```python
# Instantiate model with 100 decision trees
random_forrest_gk = RandomForestRegressor(n_estimators = 100, random_state = 42)

# Train the model
random_forrest_gk.fit(train_features_g, train_labels_g)

# Use the forest's predict method on the test data
gk_rf_predictions = random_forrest_gk.predict(test_features_g)
```

```python
#Adding other metrics for evaluation
MAE_gk_rf = mean_absolute_error(test_labels_g, gk_rf_predictions)
RMSE_gk_rf = mean_squared_error(test_labels_g, gk_rf_predictions, squared=False)
R2_gk_rf = r2_score(test_labels_g, gk_rf_predictions)

print("Metrics in predicting Final Score\n")
print("MAE:  {}".format(MAE_gk_rf))
print("RMSE: {}".format(RMSE_gk_rf))
print("R^2:  {}".format(R2_gk_rf))
```

```python
# Plot Feature Importance as barchart
x = 15
features = train_features_g.columns[:x]
importances = random_forrest_gk.feature_importances_[:x]
indices = np.flip(np.argsort(importances))

plt.figure(figsize=(12,4))
plt.bar(range(len(indices)), importances[indices], align='center')
plt.xticks(range(len(indices)), features[indices], rotation=90)
plt.show()
```

### Elasticnet Regression

```python
# Initialise the ElasticNet regressor class
elasticnet = ElasticNet(max_iter = 1000)

# Specify set of hyperparameters for performing 
params = {'alpha': [0.01, 0.1, 1, 10, 100],
          'l1_ratio': [0, 0.3, 0.6, 1]}
```

```python
# Initialize grid search
grid_search = GridSearchCV(elasticnet, params)

# Find the best set of hyperparameters
grid_search.fit(train_features_g, train_labels_g)
```

```python
# Visualise the best 3 sets of hyperparameters
cv_results = pd.DataFrame(grid_search.cv_results_)
cv_results[cv_results['rank_test_score'] <= 3].sort_values(by=['rank_test_score'])
```

```python
# Train the model using the best set of hyperparameters
elasticnet = ElasticNet(max_iter = 5000, alpha=0.1, l1_ratio=1)
elasticnet.fit(train_features_g, train_labels_g)

# Make predictions
gk_en_predictions = elasticnet.predict(test_features_g)
```

```python
# Calculate the evaluation metrics
MAE_gk_en = mean_absolute_error(test_labels_g, gk_en_predictions)
RMSE_gk_en = mean_squared_error(test_labels_g, gk_en_predictions, squared=False)
R2_gk_en = r2_score(test_labels_g, gk_en_predictions)

print("Metrics in predicting Final Score\n")
print(f"MAE:  {MAE_gk_en}")
print(f"RMSE: {RMSE_gk_en}")
print(f"R^2:  {R2_gk_en}")
```

```python
feture_importance = pd.DataFrame(list(zip(elasticnet.coef_, train_features_g.columns)), columns=['coefficient', 'variable'])
feture_importance[feture_importance['coefficient'] != 0]


```

```python
# Visualise deature importances on a graph
feture_importance = pd.DataFrame(list(zip(elasticnet.coef_, train_features_g.columns)), columns=['coefficient', 'variable'])
feture_importance = feture_importance[(feture_importance['coefficient'] < 0) | (feture_importance['coefficient'] > 0)]
feture_importance.sort_values(by=['coefficient'], ascending=False, inplace=True)

plt.figure(figsize=(13,5))

plt.bar(feture_importance['variable'], feture_importance['coefficient'])

plt.ylabel('Coefficient')
plt.xlabel('Variables')
plt.title('Feature importance')
plt.xticks(rotation = 75)
plt.show()
```

## ROI Goal Keepers

```python
# Compute predicted ROI and report metrics w.r.t to real ROI
# ROI is computed using log(price + 1) because prices do not grow linearly with the score

def ROI (prices, X_test, y_test, preds):
    
    data = pd.merge(X_test, prices, right_on=["player", "season", "matchweek"], left_on=["player", "season", "matchweek"])
    data["final_score_next_week"] = y_test.reset_index(drop=True)
    data["ROI"] = data["final_score_next_week"] / np.log(data["price"] + 1)
    data["final_score_predicted"] = preds
    data["predicted_ROI"] = data["final_score_predicted"] / np.log(data["price"] + 1)
    
    mae = mean_absolute_error(np.array(data["ROI"]), np.array(data["predicted_ROI"]))
    rmse = mean_squared_error(np.array(data["ROI"]), np.array(data["predicted_ROI"]), squared=True)
    r2 = r2_score(np.array(data["ROI"]), np.array(data["predicted_ROI"]))
    
    print("Metrics in predicting ROI\n")
    print("MAE: {}".format(mae))
    print("RMSE: {}".format(rmse))
    print("R^2: {}".format(r2))
    
    return data
```

```python
goal_keepers_rf = ROI(prices_g, X_test_g, y_test_g, gk_rf_predictions)
```

```python
goal_keepers_en = ROI(prices_g, X_test_g, y_test_g, gk_rf_predictions)
```

# Recommendations

```python
# Sort by predicted ROI and retrieve best recommendations for matchweek and role

def best_x_players (d, role, x):
    r = d[d.role == role].reset_index(drop=True)
    for suggested in range(x):
        print(suggested+1, "-", r.loc[suggested, "player"], "-", r.loc[suggested, "team"] , "(vs", r.loc[suggested, "opponent"], ") - Final Score:", (round((r.loc[suggested, "final_score_predicted"])/0.5, 0))*0.5, "- Price:", r.loc[suggested, "price"])
        print("    Real Final Score:", round(r.loc[suggested, "final_score_next_week"], 2), "- Prediction Error:", (round((r.loc[suggested, "final_score_predicted"])/0.5, 0))*0.5 - r.loc[suggested, "final_score_next_week"])
    print("")
    
    
def best_ROI (data):

    data = data.sort_values(["predicted_ROI"], ascending=False)

    for matchweek in range(19,38):
        
        d = data[data.matchweek == matchweek+1]
        print("Best ROI Predicted for Matchweek {}\n".format(matchweek+1))
        
        print("GOALKEEPERS")
        best_x_players(d, "G", 3)
        
        print("DEFENDERS")
        best_x_players(d, "D", 8)
        
        print("MIDFIELDERS")
        best_x_players(d, "M", 8)
        
        print("ATTACKERS")
        best_x_players(d, "A", 6)
        
    
    print("-----------------------------------")
```
